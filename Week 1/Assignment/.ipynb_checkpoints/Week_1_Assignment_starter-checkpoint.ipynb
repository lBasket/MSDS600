```python
import numpy as np

churn_correlation = churn_data.corr()
mask = np.triu(np.ones_like(churn_correlation, dtype=bool))
fig, ax = plt.subplots(figsize=(11,9))

sns.heatmap(churn_correlation, mask=mask, vmax=0.3, center=0, square=True)
```
